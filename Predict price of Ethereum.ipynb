```python
# import some libraries
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# Load the data
data = pd.read_csv(r'C:\Users\Yash Kumar\Documents\ETH price prediction\ETH-USD.csv')
data
```

```python
print(pd.isnull(data).sum()) 
```

```python
data = data.dropna()
```

```python
# create a variable for predicting 'n' days out into the future
projection = 14
# create a new column called prediction
data['Prediction'] = data[['Close']].shift(-projection)
# show the dataset
data
```

```python
# Create the independent dataset (X)
X = np.array(data[['Close']])
# Remove the last 14 rows data
X = X[:-projection]
print(X)
```

```python
# create the dependent dataset(y)
Y = data['Prediction'].values
Y = Y[:-projection]
print(Y)
```

```python
# split the dataset into training and test dataset
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.15)
```

```python
# create the linear model
lg = LinearRegression()
# train the model
lg.fit(x_train, y_train)
```

```python
# Test the model using score
lg_confidence = lg.score(x_test, y_test)
print('Linear Regression Confidence:', lg_confidence)
```

```python
# create a variable called x_projection and set it equal to the last 14 rows of the data from the original dataset
x_projection = np.array(data[['Close']])[-projection:]
print(x_projection)
```

```python
# print the linear regression models predictions for 14 days
lg_prediction = lg.predict(x_projection)
print(lg_prediction)
```
